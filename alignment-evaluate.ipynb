# Stochastic alignment

```python
%matplotlib inline

import pickle
from pipeline import run_pipeline
from visuals import show_alignment_result

with open("./results/preprocessed_1x.pkl", "rb") as file:
  result_1x = run_pipeline(pickle.load(file), mode=("stochastic", 1), is_processed=True)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 1, figsize=(16, 20))

show_alignment_result(axes[0], "A1", result_1x)
show_alignment_result(axes[1], "B1", result_1x)
show_alignment_result(axes[2], "C1", result_1x)

plt.show()
```
